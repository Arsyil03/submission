## Exploratory Data Analysis (EDA)

### Explore ...

```python
df_air_quality["TEMP"] = pd.to_numeric(df_air_quality["TEMP"], errors='coerce')

# Hitung rata-rata suhu per kota
city_temperature = df_air_quality.groupby("City")["TEMP"].mean().reset_index()

# Urutkan berdasarkan suhu rata-rata
city_temperature = city_temperature.sort_values(by="TEMP")

print("\nRata-rata Suhu per Kota:")
print(city_temperature)
```

**Insight:**
Hasil analisis membuktikan jika Kota Huairou merupakan kota dengan Temperature terendah dibandingkan 12 kota lainnya di dalam dataset, sedangkan kota Gucheng merupakan kota dengan temperature tertinggi dibandingkan 12 kota lainnya

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=city_temperature["TEMP"], y=city_temperature["City"], palette="Blues_r")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("City")
plt.title("Urutan Kota dengan Suhu Rata-rata Terendah ke Tertinggi")
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=city_temperature.sort_values(by="TEMP", ascending=False)["TEMP"],
            y=city_temperature.sort_values(by="TEMP", ascending=False)["City"], palette="Reds")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("City")
plt.title("Urutan Kota dengan Suhu Rata-rata Tertinggi ke Terendah")
plt.show()
```

**Insight:**
- Pada tahap Visualization pada pertanyaan pertama membuktikan bahwa kota dengan rata-rata suhu paling rendah terdapat pada kota Huairou.
- Pada tahap Visualization pada pertanyaan kedua membuktikan bahwa kota dengan rata-rata suhu paling tinggi terdapat pada kota Gucheng.

## Analisis Lanjutan (Opsional)

```python
bins = [-30, -10, 0, 10, 20, 30, 50]  
labels = ["Sangat Dingin", "Dingin", "Sejuk", "Sedang", "Hangat", "Panas"]  # Label kategori
df_air_quality["Kategori_Suhu"] = pd.cut(df_air_quality["TEMP"], bins=bins, labels=labels)
min_temp_row = df_air_quality.loc[df_air_quality["TEMP"].idxmin(), ["year", "City", "TEMP", "Kategori_Suhu"]]

print(f"Suhu terendah tercatat pada tahun {min_temp_row['year']} di kota {min_temp_row['City']} "
      f"dengan suhu {min_temp_row['TEMP']}°C yang dikategorikan sebagai '{min_temp_row['Kategori_Suhu']}'.")
```

## Insight 

Tahap ini merupakan tahap lanjutan dari program yang saya buat, kota dengan suhu terendah termasuk dalam kategori apa? apakah sangat dingin, dingin, sejuk, sedang, hangat, atau panas. Ternyata setelah dianalisis hasilnya kota dengan suhu terendah yaitu kota Huairou adalah koota dengan suhu terendah pada dari semua tahun yaitu pada tahun 2016 dengan suhu -19.9°C dengan kategori " Sangat Dingin "

## Conclusion

```python
coldest_city = city_temperature.nsmallest(1, "TEMP")
hottest_city = city_temperature.nlargest(1, "TEMP")

conclusion_1 = f"Kota dengan suhu rata-rata terendah adalah {coldest_city['City'].values[0]} ({coldest_city['TEMP'].values[0]:.2f}°C)."
conclusion_2 = f"Kota dengan suhu rata-rata tertinggi adalah {hottest_city['City'].values[0]} ({hottest_city['TEMP'].values[0]:.2f}°C)."

print("\nConclusion:")
print(f"- {conclusion_1}")
print(f"- {conclusion_2}")
```
